```python
# Imports 
import numpy as np
import random
import math 
import matplotlib.pyplot as plt
from math import factorial as fac
from numpy import sin, cos, sinh, cosh,arcsin,arccos, pi, exp,abs
%matplotlib notebook

def Choose(n,k): 
    return  fac(n)/ fac(k) / fac( n - k )

class nBezier(object):
    def __init__(self, points):
        self.points = np.array(points).astype(np.float32)

    def at(self, t):
        pt = 0 
        n= len(self.points)
        for k in range(n): 
            pt +=  Choose(n-1,k) * (1-t)**(n-1-k) * t**k * self.points[k]
        return pt
  
n = 5
# number of places to change in array
num = 5
array2 = np.reshape(np.arange(n),(n,1))

a3= np.zeros((n,1))
newArray = np.random.randint(0, n-1, size=(num, 1))
a3[newArray,:] = np.random.randint(-1, 3, size=(num, 1,1))*5
a3 = np.reshape(a3,(a3.shape[0],1))
array2 = np.hstack((array2,a3))

Curve  = nBezier(array2)

t = np.linspace(0,1,20) 

list_of_pts = []

for  item in t:
    list_of_pts.append(Curve.at(item))
    
list_of_pts = np.array(list_of_pts)


plt.scatter(array2[:,0],array2[:,1],c='r')
plt.plot(list_of_pts[:,0],list_of_pts[:,1])
#plt.savefig('Bzcurve.png')
plt.show()

'''
'''
```
