```python
# Imports 
import numpy as np
import random
import scipy
import math 
import matplotlib.pyplot as plt

from mpl_toolkits import mplot3d
from math import factorial as fac
from numpy import sin, cos, sinh, cosh,arcsin,arccos, pi, exp,abs
%matplotlib notebook

def Choose(n,k): 
    return  fac(n)/ fac(k) / fac( n - k )

class nBezier(object):
    def __init__(self, points):
        self.points = np.array(points).astype(np.float32)

    def at(self, u,v):
        pt = 0 
        n=  self.points.shape[0]
        m = self.points.shape[1]
        for k in range(n):
            for l in range(m):
                pt +=  Choose(n-1,k) * (1-u)**(n-1-k) *u**k *Choose(m-1,l) * (1-v)**(m-1-l) * v**l*self.points[k,l]
        return pt
  

# Create random points 
n = 5
# number of places to change in array
num = 15
array2 = [[x, y] for x in range(n) for y in range(n)]
array2 = np.array(array2)

a3= np.zeros((n*n,1))
newArray = np.random.randint(0, n**2-1, size=(num, 1))
a3[newArray,:] = np.random.randint(-1, 3, size=(num, 1,1))*5
a3 = np.reshape(a3,(a3.shape[0],1))
array2 = np.hstack((array2,a3))

array2 = np.reshape(array2,(n,n,3))


Curve  = nBezier(array2)
size = 20
u = np.linspace(0,1,size) 
v = np.linspace(0,1,size) 

grid_of_pts = [] 

for  item1 in u:
    pts = []
    for item2 in v:
        pts.append(Curve.at(item1,item2))
    grid_of_pts.append(pts)
        
grid_of_pts = np.array(grid_of_pts)
 
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(array2[:,:,0], array2[:,:,1], array2[:,:,2], c='r')
ax.plot_surface(grid_of_pts[:,:,0], grid_of_pts[:,:,1], grid_of_pts[:,:,2], alpha =0.8)
#plt.savefig('Bzsurf1.png')
plt.show()
'''
'''
```
