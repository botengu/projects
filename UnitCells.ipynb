```python
# Imports 
import numpy as np
import random
#import visvis as vv
import scipy
import math 
import skimage
from skimage import measure
import matplotlib.pyplot as plt
from numpy import sin, cos, sinh, cosh,arcsin,arccos, pi, exp
from scipy.spatial import Voronoi, voronoi_plot_2d, distance 

# Hexagonal structure 


# Creating a single unit cell
def polygon(center, n_sides, rho): 
    pp = [] 
    phi = pi/2 
    for i in range(int(n_sides)):
        #for  j in range(int(pi/angle)):
        #phi = 2*pi/n_sides*j
        theta = 2*pi/n_sides*i
        x = rho*sin(phi)*cos(theta)
        y = rho*sin(phi)*sin(theta)
        z = 0 # rho*cos(phi)
        pt = [center[0]+x, center[1]+y, center[2]+z]
        pp.append(pt)  
        
    pp.append(pp[0])
    pp = np.array(pp)
    plt.plot(pp[:,0],pp[:,1])

    
def unitcell1(c_x,c_y): 
    edge1 = np.array([[c_x,c_x+1],[c_y,c_y+1]]).T
    edge2 = np.array([[c_x,c_x+1],[c_y+1,c_y]]).T 
    plt.plot(edge1[:,0],edge1[:,1],color = 'k',linewidth = 6)
    plt.plot(edge2[:,0],edge2[:,1],color = 'k',linewidth = 6)
    
def unitcell2(c_x,c_y): 
    plt.plot([c_x,c_x+1,c_x+1,c_x,c_x],[c_y,c_y,c_y+1,c_y+1,c_y],color = 'k', linewidth = 6)


# Replicating the unit cell at multiple level ( first one the range is 0, second one the range is 6, then 36 etc...)
rho = 1 

N = 10
M = 5
fig = plt.figure(figsize=(2*N,2*M))
for i in range(N):
    for j in range(M):
        num = np.random.randint(0, 2)
        unitcell2(i,j)
        if(num==0):
            unitcell1(i,j)
        #else: 
            
            
            
#plt.plot([0,N,N,0,0],[0,0,M,M,0])
plt.show()
```
