```python
# This function returns the most dominant frequency being played at the prsent moement

def freque(segment,seg): # Return the frequency of a segment 
    # Review this function correctly
    # The segment seems to be affecting the graph
    Fs = 44100 
    T = 1/Fs
    t = seg
    N = Fs*t
    
    y = segment.ys
    Y_k = np.fft.fft(y)[0:int(N/2)]/N # FFT function from numpy
    Y_k[1:] = 2*Y_k[1:] # need to take the single-sided spectrum only
    Pxx = np.abs(Y_k) # be sure to get rid of imaginary part

    f = Fs*np.arange((N/2))/N; # frequency vector
    f_vec = f 
    mic_low_freq = 100
    fft_data = Pxx

    low_freq_loc = np.argmin(np.abs(f_vec-mic_low_freq))
    max_loc = np.argmax(fft_data[low_freq_loc:])+low_freq_loc
    
    return f_vec[max_loc]
    #return max_loc
```

```python
def note(fr): #finds the closest note based on the given frequency 
    global frequencies 
    
    # Find index of closest value in array 
    idx,frequ = min(enumerate(frequencies), key=lambda x: abs(fr- x[1]))

    # Find equivalent note 
    #print(Eng_notes[idx])
    return idx 
```

```python
# Array indices retrieval 

def locate(fr,notes_array):
    idx,frequ = min(enumerate(frequencies), key=lambda x: abs(fr- x[1]))

    idces = np.argwhere(notes_array == frequ)
    return idces
```
