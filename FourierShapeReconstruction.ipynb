```python
import numpy as np
import matplotlib.pyplot as plt
import random
import scipy
import math 
import os 
import cmath 
from skimage import measure
from math import sqrt
from numpy import sin, cos, sinh, cosh,arcsin,arccos, pi, exp, abs, max,min, sinh, cosh
from scipy.spatial import Voronoi, voronoi_plot_2d, distance
from matplotlib.animation import FuncAnimation


%matplotlib 


#### SHAPE CREATION

def dist(a,b):
    return sqrt((b[0]-a[0])**2+(b[1]-a[1])**2)

def divide(pts_list,num):
    finalpts_list = [] 
    length = [] 
    for i in range(pts_list.shape[0]-1):
        length.append(dist(pts_list[i,:],pts_list[i+1,:]))
    length = np.array(length)
    length = length/max(length)
    
    for i in range(pts_list.shape[0]-1):
        dens = int(num*1/length[i])
        ini = np.array([pts_list[i,:],]*dens)
        fin = np.array([pts_list[i+1,:],]*dens)
        t = np.linspace(0,1,dens)
        t = np.reshape(t,(dens,1))
        finalpts_list.append(ini+t*(fin-ini))
        
    return np.concatenate(np.array(finalpts_list)) 

pts = np.array([[1,1],[3,2],[4,5],[2,4],[1,1]]) 
num = 15

#num is the maximum amount of divisions
pts = divide(pts,num)
z = pts[:,0]+1j*pts[:,1]
N = z.shape[0]
m = np.arange(0,N)

coeff_1 = np.fft.fft(z)/N

coeff_r = coeff_1.real
coeff_i = coeff_1.imag

# Centralized shift 

coeff_r = np.hstack([coeff_r[int(N/2):N], coeff_r[0:int(N/2)]])
coeff_i = np.hstack([coeff_i[int(N/2):N], coeff_i[0:int(N/2)]])

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))
axes[0].scatter(pts[:,0],pts[:,1])
axes[0].plot(pts[:,0],pts[:,1])
axes[1].bar(m-0.25,coeff_r, width = 0.5) 
axes[1].bar(m+0.25,coeff_i, width = 0.5)
plt.show()
```

```python
# Transition function for double matrices (Complex and real)
def transiMat(MatTransi, div_num): 
    Frequencies_complex= []
    Frequencies_real= []
    shapes = [] 
    for i in range(int(div_num)):
        Curr_plt = MatTransi[0,:]+i/(div_num-1)*(MatTransi[1,:]-MatTransi[0,:])
        shapes.append(Curr_plt)
        
        coeff_r = Curr_plt.real 
        coeff_i = Curr_plt.imag
        
        coeff_r = np.hstack([coeff_r[int(N/2):N], coeff_r[0:int(N/2)]])
        coeff_i = np.hstack([coeff_i[int(N/2):N], coeff_i[0:int(N/2)]])
        
        Frequencies_real.append(coeff_r) 
        Frequencies_complex.append(coeff_i) 
    
    Freal = np.array(Frequencies_real)
    Fcomplex = np.array(Frequencies_complex)
    Shapes = np.array(shapes)
    return Shapes, Freal, Fcomplex 

# For all frequencies graphs make a figure 
def freq_Circle(arr_freq): # 3 different
    z_arr = []
    for k in range(arr_freq.shape[0]): 
        M= int(N)
        zfinal = 0
        m = np.arange(0,N)
        for i in range(M): 
            zfinal += arr_freq[k,i]*exp(2j*pi*(i)*m/N)

        zfinal = np.hstack([zfinal, zfinal[0]])
        z_arr.append(zfinal)
        
    return np.array(z_arr)


MatTransi = np.vstack([coeff_1 ,coeff_2 ]) 
shapes, real, compl = transiMat(MatTransi, 100) 

# Transform the matrix 

zfinal = freq_Circle(shapes) 

#------- 
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

def animate(i):
    axes[0].cla()
    axes[1].cla()
    axes[0].bar(m-0.25,real[i], width = 0.5) 
    axes[0].bar(m+0.25,compl[i], width = 0.5)
    axes[1].plot(zfinal[i,:].real,zfinal[i,:].imag)
    #axes[1].scatter(zfinal[i,:].real,zfinal[i,:].imag)
    
anim = FuncAnimation(fig, animate,interval=100, frames =200)
plt.show()
anim.save('fmodel4.mp4')
```
