```python
import numpy as np
import visvis as vv
import math   
import os 
import skimage
import random 

from visvis import Point, Pointset
from numpy import sin, cos, sinh, cosh, pi, exp, sqrt, min, max

# Set up mesh
# Create TPMS function 
    
def Meta2(x,y,z):
    num = 50
    n =np.random.uniform(3,27,[num,3])

    r = np.linspace(2,2, num)
    cen = np.hstack((np.array(n),np.array(r[:, None]))) 
    func = 0 
    for i in range(len(cen)): 
        func += cen[i,3]/((x-cen[i,0])**2 +(y-cen[i,1])**2 +(z-cen[i,2])**2)
    
    return func


    
def SubSphere(x,y,z):

    cen = np.array([[0,0,0,5],[0,0,2,1]])
    func = 0 

    func = cen[0,3]**2/((x-cen[0,0])**2 +(y-cen[0,1])**2 +(z-cen[0,2])**2) - cen[1,3]**2/((x-cen[1,0])**2 +(y-cen[1,1])**2 +2*(z-cen[1,2])**2) 
    
    return func

def Square(x,y,z): 
    return 3 / ( abs(x-5) + abs(y-5) +abs(z-5) )

def cube(x,y,z):
    perc = 0.8 
    a = x.shape[0]; b = x.shape[1] ; c = x.shape[2]; 

    arr = np.ones((a,b,c))
    n1 = int(perc*a/2);  n2 = int(perc*b/2); n3 = int(perc*c/2);   
    arr[(int(a/2)-n1):(int(a/2)+n1),(int(b/2)-n2):(int(b/2)+n2),(int(c/2)-n3):(int(c/2)+n3)] = 0
    return arr 

fac = 30  # Should be odd number

bot_x = -fac
top_x = fac

bot_y = -fac
top_y = fac

bot_z = -fac
top_z = fac

n = 2*fac # More of a scale then a resolution  

x = np.linspace(bot_x,top_x,n)
y = np.linspace(bot_y,top_y,n)
z = np.linspace(bot_z,top_z,n)
X, Y, Z =  np.meshgrid(x, y, z)


vol = Meta2(X,Y,Z)

# Extract a 2D surface mesh from a 3D volume (F=0)
mesh = vv.isosurface(vol,1)
m = vv.mesh(mesh)

#vv.meshWrite(os.path.expanduser('~/Desktop/CAGD/stlFiles/alone-sphere.stl'), m)
app = vv.use()
app.Run()
```
