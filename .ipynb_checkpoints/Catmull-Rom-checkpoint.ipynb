```python
# Imports 
import numpy as np
import random
import visvis as vv
import scipy
import math 
import skimage
import scipy 
from math import factorial as fac
from skimage import measure
from numpy import sin, cos, sinh, cosh,arcsin,arccos, pi, exp,abs
from visvis import Point, Pointset
from scipy.spatial import Voronoi, voronoi_plot_2d, distance
```

```python
axes = vv.gca()
def Choose(n,k): 
    return  fac(n)/ fac(k) / fac( n - k )

class spline(object):
    def __init__(self, points):
        self.points = np.array(points).astype(np.float32)

    def at(self, t):
        pt = 0 
        n= len(self.points)
        for k in range(n): 
            pt +=  Choose(n-1,k) * (1-t)**(n-1-k) * t**k * self.points[k]
        return pt
  
def create_line(pt1,pt2): 
    l = Pointset(3)
    l.append(pt1);l.append(pt2); 
    lines = vv.Line(axes,l)# Use matplot lib instead
    lines.lw = 7

    
s = 0.5

basis_mat = np.array([[-s, 2-s, s-2, s],[2*s, s-3, 3-2*s,-s],[-s, 0, s,0],[0, 1, 0,0] ]) 

# The first and last row are repeated twice
array = np.array([[0,0,0],[0,0,0],[1,2,0],[2,-3,0], [3,3,0],[4,0,0],[4,0,0] ])
# Other point [4,0,0] 

pp = Pointset(array)
l1 = vv.plot(pp, ms='.', mc='r', mw='9', ls='', mew=0 )

count = 0 

while count+4<=array.shape[0]: 
    array2 = array[count:count+4,:]
    t = np.linspace(0,1,20) 
    list_of_pts = Pointset(3)
    for item in t:
        t_mat = np.array([item**3,item**2,item,1 ])
        pt = np.matmul(t_mat, np.matmul(basis_mat,array2))
        list_of_pts.append(Point(pt))

    lines = vv.Line(axes,list_of_pts)
    lines.lw = 7

    
    count = count+1



# Curve through points (not adapted for bezier quadratic)
 



   
    
app = vv.use()
app.Run()

#array_new = scipy.interpolate.PchipInterpolator(array[:,0], array[:,1], axis=0, extrapolate=None)

#print(array_new)
```
