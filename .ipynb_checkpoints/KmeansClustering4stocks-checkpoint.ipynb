```python
import urllib
import time
import os
import numpy as np 
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from numpy.random import seed
from numpy.random import randint
from mpl_toolkits.mplot3d import axes3d
from bs4 import BeautifulSoup
from sklearn.decomposition import PCA
from sklearn.preprocessing import normalize
from matplotlib import pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
from string import ascii_lowercase
from datetime import date
```

```python
def make_soup(theurl): 
    thepage = urllib.request.urlopen(theurl)
    soup= BeautifulSoup(thepage,"html.parser")
    return soup
```

```python
# Extract all the stocks and put them in an array 
Stocks_arr=[]# two columns, one where there is the name and the other one where there is the href associated with it 
url = "https://swingtradebot.com/equities?page="
url_det = "https://swingtradebot.com/"

for i in range(1,100): 
    theurl = url + str(i)
    soup = make_soup(theurl)
    
    all_letters =  soup.find('table',{"class":"table table-striped table-bordered"}).findAll('a')
    for j in range(5,-1,-1):
        all_letters = np.delete(all_letters, j, 0)
        
    for item in all_letters:
        Stocks_arr.append([item.text,item.get('href')])
        #print(item.text)
        #print(item.text)
        #print(item.get('href'))

Stocks_arr = np.array(Stocks_arr)
# select random "even" stocks
print(Stocks_arr)
```

```python
# generate some Gaussian values
values = 2*np.random.choice(int(len(Stocks_arr)/2)-1, 30, replace=False)
```

```python
print(values)
```

```python
X_stock =  []
Symbols = [] 

# Loop over selected stocks 
for i in values: 
    # Define the ticker symbol
    tickerSymbol = Stocks_arr[i,0]
    
    # Get data on this ticker
    tickerData = yf.Ticker(tickerSymbol)

    # Get the historical prices for this ticker
    tickerDf = np.array(tickerData.history(period='1d', start='2020-6-1', end='2020-6-25'))
    if(not np.isnan(tickerDf).any()):
        tickerDf = tickerDf[:,2] / np.max(tickerDf[:,2])
        X_stock.append(tickerDf)
        print(len(tickerDf))
        Symbols.append(tickerSymbol)
# Generate np array from data 
X_stock = np.stack(X_stock, axis=0)


print(X_stock)
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X_stock)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
'''
''' 
```
