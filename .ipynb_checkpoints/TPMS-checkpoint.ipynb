```python
# Imports 
import numpy as np
import random
import visvis as vv
import scipy
import math 
import skimage
import os 
from skimage import measure
from numpy import sin, cos, sinh, cosh,arcsin,arccos, pi, exp, abs, max,min 
from visvis import Point, Pointset
from scipy.spatial import Voronoi, voronoi_plot_2d, distance 

# Spherical coordinates TPMS
def Gyroid(x,y,z):
    return (cos(x)*sin(y)+ cos(y)*sin(z)+cos(z)*sin(x))


def varGyroid(x,y,z):
    w = 1
    a = 0

    return (cos(w*x-a)*sin(w*y)+ cos(w*y)*sin(w*z)+cos(w*z)*sin(w*x-a))**2


''' 
def Gyroid2D(x,y):
    w = 0.3
    #return sin(x)*sin(y)
    return ((cos(x/w)+cos(y/w)))
    #return cos(x)*cos(y)+cos(y)+cos(x) - cos(x)*cos(y)
    #return (10*(cos(x)*cos(y)) +cos(y) + cos(x) -0.01*cos(x)*cos(y))
    #return 10*(cos(x)*cos(y)) +cos(y) + cos(x) -0.01*cos(x)*cos(y)
'''


def varGyroid2(x,y,z):
    w = 1
    t = 2
    d = 3
    a = 0
    arb_min = 0.2
    arb_max = 1
    var = x+y+z 
    var = var/(max(var)-min(a))-min(a)
    var = arb_min+(arb_max-arb_min)*var
   
    #var = (var-min(var))/(max(var)-min(var))
    #var = arb_min+(arb_max-arb_min)*var
    
    return (cos(w*x-a)*sin(t*y)+ cos(t*y)*sin(d*z)+cos(d*z)*sin(w*x-a))**2-var

def SchwarzP(x,y,z): 
    return cos(x)+ cos(y)+ cos(z)

def SchwarzD(x,y,z): 
    return sin(x)*sin(y)*sin(z) + sin(x)*cos(y)*cos(z) + cos(x)*sin(y)*cos(z) +cos(x)*cos(y)*sin(z)

def STower(x,y,z): 
    return sinh(x)+sinh(y)-sin(z)

def SSurface1(x,y,z):
    return exp(z)*cos(x) - cos(y)

def SSurface2(x,y,z):
    return sin(z+pi/2) - sinh(x)*sinh(y)

def Neovius(x,y,z):
    return 3*(cos(x)+cos(y)+cos(z))+4*cos(x)*cos(y)*cos(z)

def Schoen(x,y,z):
    return cos(x)*cos(y)+cos(y)*cos(z)+cos(z)*cos(x) - cos(x)*cos(y)*cos(z)

def Hybrid(x,y,z):# Weird
    return 10*(cos(x)*cos(y)) +cos(y)*cos(z) + cos(z)*cos(x) -0.01*cos(x)*cos(y)*cos(z)


def Square(x,y,z): 
    return 4 / ( abs(x) + abs(y) +abs(z) )

def real_Square(x,y,z): 
    return ( abs(x) + abs(y) +abs(z) -10)

def Sphere(x,y,z): 
    return (x-3)**2 + (y-3)**2 + (z-3)**2 - 36 

def cube(x,y,z):
    perc = 0.6
    a = x.shape[0]; b = x.shape[1] ; c = x.shape[2]; 
    
    arr = np.zeros((a,b,c))
    n1 = int(perc*a/2);  n2 = int(perc*b/2); n3 = int(perc*c/2);   
    arr[(int(a/2)-n1):(int(a/2)+n1),(int(b/2)-n2):(int(b/2)+n2),(int(c/2)-n3):(int(c/2)+n3)] = 1
    return arr 

def heavi(arr, thres):
    arr[arr < thres] = 0
    return arr


fac =4 # Should be odd number
n = fac*30

bot_x = -fac*math.pi
top_x = fac*math.pi 

bot_y = -fac*math.pi
top_y = fac*math.pi 

bot_z = -fac*math.pi
top_z = fac*math.pi

x = np.linspace(bot_x,top_x,n)
y = np.linspace(bot_y,top_y,n)
z = np.linspace(bot_z,top_z,n)
X, Y, Z =  np.meshgrid(x, y, z)

''' 
vol0 = Gyroid(X,Y,Z) 
vol1 = varGyroid(X,Y,Z) 
vol2 = heavi(Sphere(X,Y,Z),0) 
vol3 = cube(X,Y,Z) 
vol4 = heavi(real_Square(X,Y,Z),0)  
vol = vol1+ vol4
'''

vol = np.multiply(Gyroid(X,Y,Z),cube(X,Y,Z) )

mesh = vv.isosurface(vol,1)
m = vv.mesh(mesh)
vv.axis('off')
#vv.meshWrite(os.path.expanduser('~/Desktop/CAGD/stlFiles/losange.stl'), m)
app = vv.use()
app.Run()
```
