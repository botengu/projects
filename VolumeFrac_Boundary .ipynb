```python
import numpy as np 
import csv
import matplotlib.pyplot as plt
from numpy import sin, cos, sinh, cosh,arcsin,arccos, pi, exp, abs, max,min, sinh, cosh

r= 1
t= np.linspace(2*pi,0, 50)
x= r*sin(t)
y= r*cos(t)
fig, axes = plt.subplots(figsize=(10,10)) #plt.figure(figsize=(10,10))
axes.fill(x,y, color ='k')

def squares(): 
    den = 20
    for i in range(0,den):
        for j in range(0,den):
            x = -1+ 2*i/(den-1)
            y = -1+ 2*j/(den-1)
            #plt.plot(x,y, color ='k')
            z = (x**2+y**2)
            if(z>0.9):
                continue 
            else: 
                z = 0.12-z/12 
                pt1 = [x-0.5*z*cos(pi/4),y-0.5*z*cos(pi/4)]
                pt2 = [x-0.5*z*cos(pi/4),y+0.5*z*cos(pi/4)]
                pt3 = [x+0.5*z*cos(pi/4),y+0.5*z*cos(pi/4)]
                pt4 = [x+0.5*z*cos(pi/4),y-0.5*z*cos(pi/4)]

                pt = np.vstack([pt1,pt2,pt3,pt4,pt1])
                axes.fill(pt[:,0],pt[:,1],color ='w')

squares()        
plt.axis('equal')
plt.axis('off')
#rhino_script(List_arrays)
```
