```python
# Klein bottle

div_num = 30  
r = 5
U = np.linspace(0,pi,div_num)
V = np.linspace(0,2*pi,div_num)

u,v = np.meshgrid(U,V)


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x2 = -2/15*cos(u)*(3*cos(v)-30*sin(u) +90*cos(u)**4*sin(u)-60*cos(u)**6*sin(u)+5*cos(u)*cos(v)*sin(u))
y2 = -1/15*sin(u)*(3*cos(v)-3*cos(u)**2*cos(v)-48*cos(u)**4*cos(v)+48*cos(u)**6*cos(v) - 60*sin(u)+5*cos(u)*cos(v)*sin(u)-5*cos(u)**3*cos(v)*sin(u)- 80*cos(u)**5*cos(v)*sin(u)+80*cos(u)**7*cos(v)*sin(u))
z2 = 2/15*(3+5*cos(u)*sin(u))*sin(v)

# if you want to create a 

#ax.scatter(x1,y1,z1)

ax.plot_surface(x2, y2, z2)
     
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
x_evol = [] 
y_evol = []
z_evol = [] 


steps = 70
for i in range(steps):
    x_evol.append(x1 + i/steps*(x2 - x1))
    y_evol.append(y1 + i/steps*(y2 - y1))
    z_evol.append(z1 + i/steps*(z2 - z1))


x_evol = np.array(x_evol); y_evol = np.array(y_evol); z_evol = np.array(z_evol);  

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

def animate(i):
    plt.cla()
    ax.view_init(30, 4*i)
    ax.plot_surface(x_evol[i], y_evol[i], z_evol[i])
    
anim = FuncAnimation(fig, animate, interval=100)


ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()

#anim.save('MobiustoKlein.mp4')
```
