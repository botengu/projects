```python
# Imports 
import numpy as np
import random
import visvis as vv
import scipy
import math 
import skimage
import scipy 
from math import factorial as fac
from skimage import measure
from numpy import sin, cos, sinh, cosh,arcsin,arccos, pi, exp,abs
from visvis import Point, Pointset
from scipy.spatial import Voronoi, voronoi_plot_2d, distance
```

```python
axes = vv.gca()
def Choose(n,k): 
    return  fac(n)/ fac(k) / fac( n - k )

class nBezier(object):
    def __init__(self, points):
        self.points = np.array(points).astype(np.float32)

    def at(self, t):
        pt = 0 
        n= len(self.points)
        for k in range(n): 
            pt +=  Choose(n-1,k) * (1-t)**(n-1-k) * t**k * self.points[k]
        return pt
  
def create_line(pt1,pt2): 
    l = Pointset(3)
    l.append(pt1);l.append(pt2); 
    lines = vv.Line(axes,l)# Use matplot lib instead
    lines.lw = 7


array = np.array([[0,0,1],[1,2,3],[2,-3,5], [3,3,6], [4,0,7]])

l = vv.plot(array, ms='.', mc='r', mw='9', ls='', mew=0 )
array_int = []

# Create all intermediary points 
for i in range(array.shape[0]-1):
    array_int.append(array[i,:])
    array_int.append((array[i,:]+array[i+1,:])/2)
array_int.append(array[-1,:])    
array_int = np.array(array_int)
array_int = np.delete(array_int, 1, 0)
array_int = np.delete(array_int, -2, 0)


# Curve through points (not adapted for bezier quadratic)
''' 
for i in range(1, array_int.shape[0], 2):
    array_int[i,:] =2*array_int[i,:] - (array_int[i-1,:]+array_int[i+1,:])/2
'''
    
count = 0 

while count+2<=array_int.shape[0]: 
    Curve  = nBezier(array_int[count:count+3,:])
    t = np.linspace(0,1,10) 
    list_of_pts = [] 
    list_of_pts = Pointset(3)
    for  item in t:
        list_of_pts.append(Point(Curve.at(item)))
    lines = vv.Line(axes,list_of_pts)
    lines.lw = 4
    
    
    count = count+2
    
    
app = vv.use()
app.Run()

#array_new = scipy.interpolate.PchipInterpolator(array[:,0], array[:,1], axis=0, extrapolate=None)

#print(array_new)
```
