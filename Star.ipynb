```python
# Imports 
import numpy as np
import random
import visvis as vv
import scipy
import math 
import skimage
import os
from skimage import measure
from numpy import sin, cos, sinh, cosh,arcsin,arccos, pi, exp,abs
from visvis import Point, Pointset
from scipy.spatial import Voronoi, voronoi_plot_2d, distance 
from numpy import inf
```

```python
# Fibonacci grid 3D and metaballs
axes = vv.gca()

def cylinder(X,Y,Z,edges): 
    
    func = 0  
    #edges =np.array([[0,0,0],[1,0,0],[1,0,0],[5,5,0]]) 

    for i in range(len(edges)): 
        if(i%2==0):
            p1 = edges[i,:]
            p2 = edges[i+1,:]
            vec = p2 - p1
            a,b,c = vec
            xx, yy,zz = p1+vec 
            coeff = np.linalg.norm(p2-p1)
            R =0.1
            R *=coeff
            cyl = R**2/((a*(Y-yy)-b*(X-xx))**2+(c*(X-xx)-a*(Z-zz))**2+(b*(Z-zz)-c*(Y-yy))**2)
            sph = coeff**2/((X-xx)**2 + (Y-yy)**2 + (Z-zz)**2)
            func += np.multiply(removeinf(cyl), heavi(sph,2))

    return func 

def Metaballs(X,Y,Z): 
    cen = np.array([[0,0,0,1]])
    return cen[0,3]/((X-cen[0,0])**2 +(Y-cen[0,1])**2 +(Z-cen[0,2])**2)
    

def heavi(arr, thres):
    arr[arr < thres] = 0
    arr[arr > thres] = 1
    return arr

def removeinf(arr):
    arr[arr == inf] = 4
    return arr

def create_line(pt1,pt2): 
    l = Pointset(3)
    l.append(pt1);l.append(pt2); 
    vv.Line(axes,l)



fac = 10  # Should be odd number

bot_x = -fac
top_x = fac

bot_y = -fac
top_y = fac

bot_z = -fac
top_z = fac

n = 20*fac # More of a scale then a resolution  

x = np.linspace(bot_x,top_x,n)
y = np.linspace(bot_y,top_y,n)
z = np.linspace(bot_z,top_z,n)
X, Y, Z =  np.meshgrid(x, y, z)



        
vert = Point(0.0,0.0,0.0)
min_rad = 3 
max_rad = 5

edges = [] 
G = (1 + 5 ** 0.5) / 2
N = 30
pp = Pointset(3)
for n in range(0,N):
    Phi_n = arccos(1 - 2*n/N)
    Theta_n = 2*pi * n / G
    rho = 1/ (random.uniform(1/max_rad, 1/min_rad))
    x_n = vert[0] + rho*sin(Phi_n)*cos(Theta_n) 
    y_n = vert[1] + rho*sin(Phi_n)*sin(Theta_n)
    z_n = vert[2] + rho*cos(Phi_n)
    vert_new= Point(x_n,y_n,z_n)
    edges.append([0,0,0])
    edges.append([x_n,y_n,z_n])
    create_line(vert,vert_new)
    
edges = np.array(edges)


app = vv.use()
app.Run()
```

```python
vol = cylinder(X,Y,Z,edges) + Metaballs(X,Y,Z)

# Extract a 2D surface mesh from a 3D volume (F=0)
mesh = vv.isosurface(vol,0.1)
m = vv.mesh(mesh)
vv.meshWrite(os.path.expanduser('~/Desktop/CAGD/stlFiles/star3.stl'), m)
'''
''' 

app = vv.use()
app.Run()
```
